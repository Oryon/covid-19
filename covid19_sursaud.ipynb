```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn')
```

```python
data = pd.read_csv("https://www.data.gouv.fr/fr/datasets/r/eceb9fb4-3ebc-4da3-828d-f5939712600a", sep=',')
data = data.fillna(0)
data
```

```python
data.dtypes
```

```python
metadata = pd.read_csv("https://www.data.gouv.fr/fr/datasets/r/2265d880-5260-4b34-8827-b3d76180032e", sep=";").drop(['Unnamed: 1', 'Unnamed: 3'], axis=1)
metadata
```

```python
data_jour = data.groupby(['date_de_passage']).sum()
data_jour
```

```python
fig, ax = plt.subplots(figsize=(14,7))
#plt.plot(data_jour.index, data_jour['nbre_pass_tot'], label='Passage tous', linewidth=2, marker='o')
plt.plot(data_jour.index, data_jour['nbre_pass_corona'], label='Passage covid', linewidth=2, marker='o')
plt.plot(data_jour.index, data_jour['nbre_hospit_corona'], label='Hospitalisation covid', linewidth=2, marker='o')
plt.legend(fontsize='x-large')
plt.annotate('<- Confinement', (21,10000), color='C0', fontsize='x-large', fontweight='bold')
plt.axvline(21, color='C0', linestyle='dashed')
plt.xticks(data_jour.index, data_jour.index, rotation='vertical')
ax.set_title('COVID-19: Passage aux urgences sur suspicion de covid', fontsize='x-large')
plt.show()
```

```python
data_age = data.groupby(['sursaud_cl_age_corona']).sum().reset_index().drop(0)
data_age
```

```python
age_group = pd.read_csv("https://www.data.gouv.fr/fr/datasets/r/cbed61a8-64e8-4c8c-8e97-65d3a31a99b6", sep=";")
age_group
```

```python
fig, ax = plt.subplots(figsize=(14,7))
plt.plot(age_group['Unnamed: 1'][1:], data_age['nbre_hospit_corona'] / data_age['nbre_pass_corona'], label='Ratio d\'hostpitalisation pour covid', linewidth=2, marker='o')
plt.legend(fontsize='x-large')
ax.set_title('COVID-19: Taux hospitalisation par âge par rapport au nombre de visite aux urgences', fontsize='x-large')
plt.show()
```
